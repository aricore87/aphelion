(Q) Simulate motion of planet around the Sun. Take precise values in SI units

Mass of the Sun = 1.99E30 kg
Mass of planet = 6.417E23 kg
Distance of planet from the Sun = 2.094E11 meters

Solve in XY plane.

1) Give sufficient speed to get circular orbit.
2) Determine the period of revolution by checking when does it come to original position. You may use trial and error.
3) Change the speed by 5% and check how does the orbit change. Do it for +5% and -5% change.
Determine from simulation how does the distance from the Sun changes when it goes opposite to the Sun in each case.
Also determine from your simulation how does the period of revolution change in each case

```python
import numpy as np

# Constants
G = 6.67430e-11
M_SUN = 1.99e30
R0_VAL = 2.094e11
DT = 86400  # 1 day
STEPS = 2000

def run_sim(v_init):
    # position and velocity
    r = np.array([R0_VAL, 0.0])
    v = np.array([0.0, v_init])

    #trackers
    initial_r_mag = np.linalg.norm(r)
    opp_dist = 0
    period = 0

    for i in range(1, STEPS):
        #1. calculate acceleration vector: a = -GM / |r|^3 * r
        r_mag = np.linalg.norm(r)
        accel = -(G * M_SUN / r_mag**3) * r

        #2. update velocity and position
        v += accel * DT
        r += v * DT

        #checking half-orbit (y crosses from + to -)
        if i > 10 and r[1] < 0 and (r[1] - v[1]*DT) >= 0:
            opp_dist = np.linalg.norm(r)

        #checking for full orbit (y crosses from - to +)
        if i > 100 and r[1] > 0 and (r[1] - v[1]*DT) <= 0 and period == 0:
            period = i * DT

    return period / 86400, opp_dist

#initial circular speed
v_circ = np.sqrt(G * M_SUN / R0_VAL)

print(f"{'Condition':<15} | {'Period (Days)':<15} | {'Opp. Distance (m)':<15}")
print("-" * 50)

for label, factor in [("Circular", 1.0), ("-5% Speed", 0.95), ("+5% Speed", 1.05)]:
    p, d = run_sim(v_circ * factor)
    print(f"{label:<15} | {p:<15.1f} | {d:<15.2e}")
```
